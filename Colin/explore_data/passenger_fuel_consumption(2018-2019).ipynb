## Passenger Data (2018-2019)

```python
passenger = pd.read_csv("D:\Python(New)\Project\Mid_Term_Project\Colin\original\passenger_2018_2019.csv")
```

```python
passenger.info()
```

```python
passenger.describe()
```

```python
passenger_data = passenger[['unique_carrier','airline_id','unique_carrier_name','origin_airport_id','origin_city_market_id','origin','origin_city_name','origin_country','origin_country_name','dest_airport_id','dest_city_market_id','dest','dest_city_name','dest_country','dest_country_name', 'aircraft_group','aircraft_type','seats','passengers','freight','mail','distance','ramp_to_ramp','air_time','month','year','distance_group']]
passenger_data
```

```python
col = ['airline_id','origin_airport_id','origin_city_market_id','dest_airport_id','dest_city_market_id', 'aircraft_group', 'aircraft_type','distance_group','month','year']
passenger_data[col] = passenger_data[col].astype('str')
passenger_data['fl_date'] = passenger_data[['month', 'year']].agg('/'.join, axis=1)
passenger_data = passenger_data.drop(columns = ['month', 'year'])
passenger_data
```

```python
passenger_data.to_csv('D:\Python(New)\Project\Mid_Term_Project\Colin\clean\passenger_data_2018_2019.csv', index = False)
```

## Fuel Consumption (2018-2019)

```python
fuel_consump = pd.read_csv("D:\Python(New)\Project\Mid_Term_Project\Colin\original\\fuel_consumption_2018_2019.csv")
fuel_consump
```

```python
fuel_data = fuel_consump[['airline_id','unique_carrier','carrier_name','sdomt_gallons','satl_gallons','spac_gallons','slat_gallons','sint_gallons','ts_gallons','tdomt_gallons','tint_gallons','total_gallons','sdomt_cost','satl_cost','spac_cost','slat_cost','sint_cost','ts_cost','tdomt_cost','tint_cost','total_cost','month','year']]
```

```python
cols = ['airline_id','month','year']
fuel_data[cols] = fuel_data[cols].astype('str')
fuel_data['date_time'] = fuel_data[['month', 'year']].agg('/'.join, axis=1)
fuel_data = fuel_data.drop(columns =['month','year'])
fuel_data
```

```python
fuel_data.to_csv('D:\Python(New)\Project\Mid_Term_Project\Colin\clean\\fuel_data_2018_2019.csv', index = False)
```
